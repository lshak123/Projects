````python
targetDir = 'D:\\프로젝트 정리\\시각 장애인 봇 서비스(Conference)\\Image Captioning Dataset (인도 보행 영상)\\인도보행 영상\\바운딩박스\\Bbox_2_total'

# targetDir에서 .xml파일 이름들 리스트로 가져오기
file_list = os.listdir(targetDir)
xml_list = []
for file in file_list:
    if '.xml' in file:
        xml_list.append(file)
````

### xml을 dataframe으로 저장 및 전처리

````python
cd ../Bbox_2_total
````

````python
anno_total = pd.DataFrame({'image':'ZED3_KSC_129930_L_P034801.png', 'occluded':1, 
                           'ybr':499.66, 'xbr':420.20, 'ytl':385.64, 'xtl':186.24, 
                           'label':'car', 'z_order':0.0}, index=[0])

for k in xml_list:

    tree = ET.parse(k)
    root = tree.getroot()

    image_names = []

    for i in root.findall('image'):
        for j in i:
            image_name = i.attrib['name']
            image_names.append(image_name)

    image = pd.DataFrame(image_names).rename(columns={0:'image'})
    box = pd.read_xml(k, xpath='image/box')

    anno = pd.concat([image, box], axis=1)

    anno_total = pd.concat([anno_total, anno], axis=0)
````

````python
anno_total = anno_total.reset_index(drop=True)[1:].reset_index(drop=True)
anno_total = anno_total[['image', 'ybr', 'xbr', 'ytl', 'xtl', 'label']] ; anno_total
````

````python
anno_total = anno_total.rename(columns={"image":"filename", "xtl":"x1", "ytl":"y1", 
                                        "xbr":"x2", "ybr":"y2", "label":"labels"}) ;anno_total
````

````python
anno_total['width'] = 1920
anno_total['height'] = 1080
````

````python
anno_total = anno_total[['filename', 'width', 'height', 'x1', 'y1', 'x2', 'y2', 'labels']] ;anno_total
````

````python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

anno_total.labels = encoder.fit_transform(anno_total.labels)
````

````python
print('classes:', encoder.classes_, '\n')
````

````python
label_classes = pd.DataFrame({'names':encoder.classes_, 
                              'numbers':sorted(anno_total.labels.unique().tolist())}) ;label_classes
````

````python
rep_list = anno_total.groupby('filename')[['x1']].count()['x1'].tolist()
print(rep_list, end='\\')
````

````python
# 새로운 annotation 구조

data = []

i=0
for rep in rep_list:  
    annotation = dict(bboxes = np.array(anno_total.iloc[i : i+rep, 3:7], dtype=np.float32).reshape(-1, 4), 
                      labels = np.array(anno_total.iloc[i : i+rep, 7], dtype=np.compat.long))
    total = dict(filename = anno_total.filename[i], width=1920, height=1080, ann=annotation)
    data.append(total)
    i += rep
````
